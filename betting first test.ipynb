```python
import numpy as np
import sqlite3

conn = sqlite3.connect("soccer_all_leagues.sqlite")
c = conn.cursor()


bankroll = 1000
betsize = 1

def  placebet(bet,odds,betsize,g_home,g_away):
    if bet == 1:
        if g_home > g_away:
            deltabankroll = (1/odds)*betsize - betsize
        else:
            deltabankroll = - betsize
    if bet == 'X':
        if g_home == g_away:
            deltabankroll = (1/odds)*betsize - betsize
        else:
            deltabankroll = -betsize
    if bet == 2:
        if g_home < g_away:
            deltabankroll = (1/odds)*betsize - betsize
        else:
            deltabankroll = -betsize
    return [deltabankroll]
    
def ratingodds(p_fav):
    draw = -1.2463990693*(p_fav**2)+(1.23969507828*p_fav) - (0.00557651949082)
    win_fav = p_fav - 0.5 *draw
    win_dog = 1-p_fav - 0.5 * draw
    return [win_fav,draw,win_dog]

def oddstopercentage(odds):
    bookiep = [None]*3
    bookiep[0] = 1/odds[0][0]
    bookiep[1] = 1/odds[0][1]
    bookiep[2] = 1/odds[0][2]
    return [bookiep]
#count number of matches to use for loop later

c.execute('SELECT Count(*) FROM Match')
matchcount = c.fetchall()
matchcount= int(matchcount[0][0])
print(matchcount)



for i in range(matchcount - 5000):
    var = i + 5000
    p = 50
    print(var)
    deltabankroll = 0

#match info
    
    c.execute('''SELECT B365H, B365D, B365A
          FROM match ORDER BY DATE ASC LIMIT 1 OFFSET ?  ''', (var,))
    
    bookieodds = c.fetchall()
    if bookieodds[0][0] is None:
        continue
    c.execute('''SELECT home_team_goal
          FROM match ORDER BY DATE ASC LIMIT 1 OFFSET ?  ''', (var,))
    g_home = c.fetchall()    

    c.execute('''SELECT away_team_goal
          FROM match ORDER BY DATE ASC LIMIT 1 OFFSET ?  ''', (var,))
    g_away = c.fetchall()

    c.execute('''SELECT home_team_score, away_team_score, home_field_adv
          FROM match ORDER BY DATE ASC LIMIT 1 OFFSET ?  ''', (var,))
    
    scores = c.fetchall()
    
    home_team_score = scores[0][0]
    away_team_score = scores[0][1]
    home_field_adv = scores[0][2]

    p_home = 1 / (1 + np.exp((-1)*(home_team_score + home_field_adv - away_team_score)))
    p_away = 1 - p_home
    
    if p_home > p_away:
        ratingp = draw(p_home)
    else:
        ratingptemp = draw(1-p_home)
        ratingp[0] = ratingptemp[2]
        ratingp[1] = ratingptemp[1]
        ratingp[2] = ratingptemp[0]
    
    bookiep = oddstopercentage(bookieodds)
   
    for i in range(len(ratingp)):
        if ratingp[i] > bookiep[0][i]:
            if i == 0:
                bet = 1
            if i == 1:
                bet = 'X'
            else:
                bet = 2
            deltabankroll = placebet(bet,bookiep[0][i],betsize,g_home,g_away)
            deltabankroll = deltabankroll[0]
    bankroll += deltabankroll
    print(bankroll)
```
